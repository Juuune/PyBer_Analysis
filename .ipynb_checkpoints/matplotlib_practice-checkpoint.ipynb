```python
import matplotlib.pyplot as plt
%matplotlib inline
```

## Line Chart

```python
# Set the x-axis to a list of strings for each month.
x_axis=["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
# Set the y-axis to a list of floats as the total fare in US dollors accumulated for each month 
y_axis=[10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot with ax.plot()
fig, ax = plt.subplots()
# Create the plot
ax.plot(x_axis, y_axis, marker='D', color='green',linewidth= 2, label = 'Boston')
# Create labels for the x and y axes.
ax.set_xlabel("Date")
ax.set_ylabel('Fare($)')
# Set the y limit between 0 and 45. 
ax.set_ylim(0,45)
# Create a title 
ax.set_title("Pyber Fare by Month")
# Add a grid
ax.grid()
# Add the legend
ax.legend()

# Save Image 
# plt.savefig("PyberFarebyMonth_example.png")

plt.show()
```

```python
# Create the plot with plt.plot()
# Create the plot
plt.plot(x_axis, y_axis, marker='D', color='green',linewidth= 2, label = 'Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel('Fare($)')
# Set the y limit between 0 and 45. 
plt.ylim(0,45)
# Create a title 
plt.title("Pyber Fare by Month")
# Add a grid
plt.grid()
# Add the legend
plt.legend()
# Save Image 
# plt.savefig("PyberFarebyMonth_example2.png")
plt.errorbar(x_axis,y_axis)
plt.show()
```

## Bar Chart

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot 
plt.bar(x_axis, y_axis, color ='green', label ='Boston')
# Create labels for x and y axes
plt.xlabel('Date')
plt.ylabel('Fare($)')
# Create a title 
plt.title("Pyber Fare by Month")
# Add the legend
plt.legend()
```

```python
# Create the horizontal bar plot
plt.barh(x_axis,y_axis, color = 'magenta', label = 'Boston' )
plt.xlabel('Fare($)')
plt.ylabel('Date')
plt.title('Pyber Fare by Month')
plt.legend()
# Invert y-axis order
plt.gca().invert_yaxis()
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis)
```

```python
# Create the barh plot with ax.plt()
fig, ax =plt.subplots()
ax.barh(x_axis, y_axis, color='cyan', label='Chicago')
ax.set_xlabel('Fare($)')
ax.set_ylabel('Date')
ax.set_title('Pyber Fare by Month')
plt.gca().invert_yaxis()
ax.legend()
```

## Scatter Plot


```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create scatter plot with plt.plot()
plt.plot(x_axis, y_axis, 'o')
```

```python
# Add annoatation to scatter plot 
plt.scatter( y_axis,x_axis, color= 'red', label ='Chicago')
plt.xlabel('Fare($)')
plt.ylabel('Date')
plt.title('Pyber Fare by Month')
plt.gca().invert_yaxis()
plt.legend()
```

```python
# Crate bubble chart with for loop
y_axis_larger =[]
for data in y_axis:
    y_axis_larger.append(data*3)
    
plt.scatter(x_axis, y_axis,s=y_axis_larger)
```

```python
# Increase y_axis size with list comprehension 
plt.scatter(x_axis,y_axis, s=[i*3 for i in y_axis])
```

```python
# Create ax.plt() add skill drill requirements
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, color='skyblue',
           linewidth= 2, edgecolor='black',
           s= [i*5 for i in y_axis],
           alpha = 0.8,
# alpha makes dots on plot transparent with indicated degree
          label='Boston')
ax.set_xlabel('Fare($)')
ax.set_ylabel('Date')
ax.set_title('Pyber Fare bt Month')
plt.gca().invert_yaxis()
ax.legend()
ax.set_xlim(0,50)
```

## Pie Chart

```python
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
# Assign 12 colors, one for each month
colors =["slateblue","magenta","lightblue", "green","yellowgreen","greenyellow"
        ,"yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
# Pie chart explode
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
# resize pie chart
plt.subplots(figsize=(8,8))
plt.pie(y_axis, 
        explode=explode_values, 
        colors = colors,
        labels=x_axis, 
        autopct='%.1f%%')
plt.show()
```

```python
# Create pie chart with ax.plot
# Assign 12 colors, one for each month
colors =["slateblue","cornflowerblue","lightblue", "olivedrab","yellowgreen","greenyellow"
         ,"gold", "orange", "lightyellow", "indianred", "lightsalmon", "mistyrose"]
# Pie chart explode
explode_values = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
fig, ax = plt.subplots(figsize=(8,8))
# resize pie chart

ax.pie(y_axis, labels=x_axis,
       colors=colors,
       explode = explode_values,
       autopct='%.1f%%',
       shadow = True, 
       startangle=-95)
# Convert orders
plt.gca().invert_yaxis()

plt.show()
```
